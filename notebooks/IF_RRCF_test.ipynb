```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
import time
import rrcf
sns.set_style("whitegrid")
```

```python
IF = IsolationForest(n_estimators = 100,
                     max_samples  = 256,
                     contamination= 'auto',
                     behaviour='new',
                     random_state=42)
```

```python
data = np.load("taxi_all.npz")

parmas = {"trees"      : 100,
         "sample_size" : 1000,
         "data_dim"    : 48}

D = np.zeros((len(data["passengers"]), 48))
for i in range(48, len(data["passengers"])):
    D[i,:] = data["passengers"][i-48:i]
```

```python
IF.fit(D)
```

```python
score = IF.score_samples(D)
```

```python
plt.plot(-score)
```
